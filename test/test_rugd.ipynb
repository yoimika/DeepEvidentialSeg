# Test RUGD dataset
用来看看 h5 数据弄得怎么样

```python
import h5py
import numpy as np
import cv2
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
from utils import visualize_with_legend
```

```python
file = h5py.File('../data/test_test.h5', 'r')
file.keys()
```

```python
idx = np.random.randint(0, len(file['images']))

image, label = file['images'][idx], file['labels'][idx]

image = cv2.imdecode(image, cv2.IMREAD_COLOR_RGB)
label = cv2.imdecode(label, cv2.IMREAD_UNCHANGED)

print(f'Image shape: {image.shape}, Label shape: {label.shape}')

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image)
ax[0].set_title('Image')
visualize_with_legend(ax[1], label, 'label')
plt.show()
```
